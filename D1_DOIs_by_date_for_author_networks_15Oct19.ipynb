## Using the FinalDOIs, with publication date defined as when the DOI was first created

###  Sorting DOIs into 3 sections after realisation that the publication rate was very similar over the course of Oxford BRC2 

### These 3 DOI sets are then exported for use in VOSviewer to create co-authorship networks for comparison of start, midpoint and end of the BRC

```python
import pandas as pd
```

```python
df =pd.read_csv('./C1in_2366_DOIsWith_2counts_and_year(15Oct19).csv',
                    index_col=['pub_date_CR_API'], parse_dates=True, converters={'add_api':pd.io.json.json_normalize})
```

```python
df.sort_index(inplace=True)
df.head()
```

```python
df.index.dropna().size /3
#of 2361 rows with dates, split into 3
```

```python
df_start =df[0:788]
                 
df_mid =df[788:1576]
df_end =df[1576:2366]
```

```python
df_start['FinalDOI'].to_csv('./author_networks/dois_Start.csv', index=False, header=False)
df_mid['FinalDOI'].to_csv('./author_networks/dois_Mid.csv', index=False, header=False)
df_end['FinalDOI'].to_csv('./author_networks/dois_End.csv', index=False, header=False)
```
