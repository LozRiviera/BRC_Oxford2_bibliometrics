## Using the FinalDOIs, with publication date defined as when the DOI was first created

###  Sorting DOIs into 3 sections after realisation that the publication rate was very similar over the course of Oxford BRC2 

### These 3 DOI sets are then exported for use in VOSviewer to create co-authorship networks for comparison of start, midpoint and end of the BRC

```python
import pandas as pd
```

```python
df_bod =pd.read_csv('./C2out_2367_DOIsWith_2counts_and_year(08Aug19).csv', index_col=['pub_date_CR_API'], parse_dates=True)
```

```python
df_bod.sort_index(inplace=True)
df_bod.head()
```

```python
df_bod.count()
```

```python
df_bod.index.dropna().size /3
#of 2362 split into 3
```

```python
start_time = df_bod.index[0]
```

```python
start_time
```

```python
end_time = df_bod.sort_index(ascending=False).index[0]
```

```python
end_time
```

```python
time = end_time-start_time
```

```python
time /3
```

```python
df_start =df_bod[0:788]
                 
df_mid =df_bod[788:1576]
df_end =df_bod[1576:2363]
```

```python
df_start['FinalDOI'].to_csv('./years_for_Gephi/dois_Start.csv', index=False, header=False)
df_mid['FinalDOI'].to_csv('./years_for_Gephi/dois_Mid.csv', index=False, header=False)
df_end['FinalDOI'].to_csv('./years_for_Gephi/dois_End.csv', index=False, header=False)
```
